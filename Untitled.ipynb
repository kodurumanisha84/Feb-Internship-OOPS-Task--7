```python
#Problem 1: Smart Light Controller
class SmartHomeLight: 
  def __init__(self,light_name):
    self.light_name=light_name
    self.status="off"
  def Turn_ON(self):
    self.status="ON"
    print(self.light_name," is ON")
  def Turn_OFF(self):
    self.status="OFF"
    print(self.light_name,"is OFF")
light=SmartHomeLight("Bedroom Light")
light.Turn_ON()
    
    
```

```python
#Problem 2: Employee ID Card System
class Employee:
    def __init__(self,name,id,department):
        self.name=name
        self.id=id
        self.department=department
    def display(self):
        print("Employee ID Card")
        print("Employee Name:",self.name)
        print("Employee Id:",self.id)
        print("Department:",self.department)
details=Employee("Rahul","EMP102","IT")
details.display()
```

```python
#Problem 3: Mobile Contact Record
class Mobile:
    def __init__(self,name,phone_number):
        self.name=name
        self.phone_number=phone_number
    def display(self):
         print("Contact Saved")
         print("Name:",self.name)
         print("Phone:",self.phone_number)
Contact_INFO=Mobile("Anita", 9876543210)
Contact_INFO.display()
         
```

```python
#Problem 4: Product Price Tag Generator
class Product:
    def __init__(self,product_name,price):
        self.product_name=product_name
        self.price=price
    def display(self):
        print("Product:",self.product_name)
        print("Price:",self.price)
details=Product("HeadPhones","Rs.2499")
details.display()
```

```python
#Problem 5: Movie Rating Display System
class MovieRating:
    def __init__(self,movie_name,rating):
        self.movie_name=movie_name
        self.rating=rating
    def display(self):
        print("Movie:",self.movie_name)
        print("Rating:",self.rating,"/5")
movieDetails=MovieRating("Inception",4.8)
movieDetails.display()
        
```

```python
#Problem 6: Delivery Address Manager
class Delivery:
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def display(self):
        print("Delivery Details")
        print("Name:",self.name)
        print("Address:",self.address)
Delivery_Details=Delivery("Suman","Hyderabad")
Delivery_Details.display()
        
```
